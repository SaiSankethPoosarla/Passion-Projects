```python
#Loop for all required playoffs/splits, dataframes will be combined at the end

all_teamstats11 = []
all_playerstats11 = []
algs_page_url = "https://apexlegendsstatus.com/algs/"

playerp11 = []
playerimg11 = []
playersrc11 = []
teamp11 = []
teamimg11 = []
teamsrc11 = []
```

```python
import time
for l in algs_links:
    
    competition_name = l.split('/')[4]+" "+l.split('/')[5]
    
    data = requests.get(l)
    soup = BeautifulSoup(data.text)
    playerStats_table = pd.read_html(data.text, attrs = {'id': 'playersStatsTable'})
    teamstats_table = pd.read_html(data.text, attrs = {'id': 'teamsStatsTable'})
    playerstats_table = playerStats_table[0]
    teamstats_table = teamstats_table[0]
    
    playertable = soup.find('table',id="playersStatsTable") #using beautiful soup to add a src column, will be used to join team and player tables
    playerp11 = playertable.find_all('p', class_='teamImageContainer')
    playerimg11 = [p.find('img') for p in playerp11]
    playersrc11 = [p.get('src') for p in playerimg11]

    teamtable = soup.find('table',id="teamsStatsTable")
    teamp11 = teamtable.find_all('p', class_='teamImageContainer')
    teamimg11 = [p.find('img') for p in teamp11]
    teamsrc11 = [p.get('src') for p in teamimg11]
    
    playerstats_table["Competition"] = competition_name
    playerstats_table["Team_src"] = playersrc11
    teamstats_table["Competition"] = competition_name
    teamstats_table["Team_src"] = teamsrc11
    all_teamstats11.append(teamstats_table)
    all_playerstats11.append(playerstats_table)
    time.sleep(2)
    
```

```python
teamstats_combined11_df = pd.concat(all_teamstats11)
playerstats_combined11_df = pd.concat(all_playerstats11)
```
