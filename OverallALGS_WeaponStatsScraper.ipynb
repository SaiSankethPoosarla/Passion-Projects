```python
### get required competition links
start_url = "https://apexlegendsstatus.com/algs/"
data = requests.get(start_url)
soup = BeautifulSoup(data.text)

### finds all ALGS aka LAN tournaments only. may change to include pro league later

comp_links = soup.find_all('a',class_="algsRegionElem linkNoStyle")
comp_links = [l.get("href") for l in comp_links]
comp_links = [l for l in comp_links if '/ALGS-' in l]
comp_links = [f"https://apexlegendsstatus.com{l}" for l in comp_links]

### create list of links corresponding to the weapons page for each competition

comp_links = [f"{l}#tab-weaponsStats" for l in comp_links]
```

```python
weaponStats = []
```

```python
weaponTable = []
weaponStats_comb = []
driver = webdriver.Edge()
time.sleep(random.randint(1,3))

for l in comp_links:
    
    competition_name = l.split('/')[4]+" "+l.split('/')[5] #get competition name
    
    driver.get(l)                      #get weapon tab
    time.sleep(random.randint(1,3))
    ActionChains(driver)\
        .scroll_by_amount(0, 150)\
        .perform()
    time.sleep(random.randint(1,3))
    weapondropdownclick = driver.find_element('id','dropdownMenuButton')
    weapondropdownclick.click()   #find the dropdown button and open it
    time.sleep(random.randint(1,3))
    #create the list of dropdown items
    
    playerdropdownitems = driver.find_element('id','weaponsSearch')
    playerdropdownitems2 = playerdropdownitems.find_elements(By.TAG_NAME,'a')
    
    for i in range(0,len(playerdropdownitems2)):

        time.sleep(random.randint(1,2))
        playerdropdownitems2[i].click()
        time.sleep(random.randint(1,2))
        weapondropdownclick.click()
        time.sleep(random.randint(1,2))

        # use selenium to grab the weapon tables here


        seleniumtable = driver.find_element(By.ID, 'weaponsStatsTable').get_attribute("outerHTML")
        weaponTable = pd.read_html(seleniumtable)
        weaponTable = weaponTable[0]
        
        # need to add player name and competition name columns to weapon table
        
        weaponTable["Player"] = playerdropdownitems2[i].text
        weaponTable["Competition"] = competition_name
        
        weaponStats.append(weaponTable)
        ###

        time.sleep(random.randint(1,2))
        x = x+1
        delta_y = 40 * x
        ActionChains(driver)\
            .scroll_by_amount(0, delta_y)\
            .perform()
        time.sleep(0.5)
        
        if x % 10 == 0:
            driver.refresh()
            time.sleep(2)
            ActionChains(driver)\
            .scroll_by_amount(0, 300)\
            .perform()
            time.sleep(2)
            weapondropdownclick = driver.find_element('id','dropdownMenuButton')
            weapondropdownclick.click()
            time.sleep(1)
            playerdropdownitems = driver.find_element('id','weaponsSearch')
            playerdropdownitems2 = playerdropdownitems.find_elements(By.TAG_NAME,'a')
            ActionChains(driver)\
            .scroll_by_amount(0, delta_y)\
            .perform()

            continue
```
